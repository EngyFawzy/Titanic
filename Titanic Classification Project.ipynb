```python
train_set= train_set.dropna(subset=["Embarked"])
test_set = test_set.dropna(subset=["Embarked"])
```

```python
train_set = train_set.drop("Cabin", axis=1)
test_set = test_set.drop("Cabin", axis=1)
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()

embarked = train_set[["Embarked"]]
embarked_test = test_set[["Embarked"]]
embarked_1hot = encoder.fit_transform(embarked)
embarked_test_1hot = encoder.fit_transform(embarked_test)
matrix = embarked_1hot.toarray()
matrix1 = embarked_test_1hot.toarray()
```

```python
matrix
```

```python
column_names= ["C","Q","S"]
for i in range (len(matrix.T)):
        train_set[column_names[i]] = matrix.T[i]
for j in range (len(matrix.T)):
        test_set[column_names[j]] = matrix1.T[j]
```

```python
gender = train_set[["Sex"]]
gender_test = test_set[["Sex"]]
gender_1hot = encoder.fit_transform(gender)
gender_test_1hot = encoder.fit_transform(gender_test)
matrix2= gender_1hot.toarray()
matrix3= gender_test_1hot.toarray()
```

```python
matrix2
```

```python
column_names2= ["Female", "male"]
for i in range (len(matrix2.T)):
        train_set[column_names2[i]] = matrix2.T[i]
for j in range (len(matrix3.T)):
        test_set[column_names2[j]] = matrix3.T[j]
```

```python
train_set = train_set.drop("Embarked", axis=1)
train_set = train_set.drop("Sex", axis=1)
test_set = test_set.drop("Embarked", axis=1)
test_set = test_set.drop("Sex", axis=1)
train_set
```

```python
train_data = train_set.drop("Survived", axis=1)
train_label = train_set["Survived"]
```

```python
train_data.info()
```

```python
test_set.info()
```

```python
test_set = test_set.dropna(subset=["Fare"])
ID = test_set["PassengerId"]
test_set = test_set.drop("PassengerId", axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
train_data_scaled = scaler.fit_transform(train_data)
train_label_numpy = train_label.to_numpy()
```

```python
from sklearn.ensemble import GradientBoostingClassifier as gb
gradient=gb(learning_rate = 0.01, max_depth = 12, random_state = 0).fit(train_data , train_label)
```

```python
predict = gradient.predict(test_set)
```

```python
output = pd.DataFrame({'PassengerId':ID,'Survived':predict})
output
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score

clf = KNeighborsClassifier(n_neighbors = 13)
scoring = 'accuracy'
score = cross_val_score(clf, train_data, train_label, cv=10, n_jobs=1, scoring=scoring)
score
```

```python
round(np.mean(score)*100, 2)
```

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier()
score = cross_val_score(clf, train_data, train_label, cv=10, n_jobs=1, scoring=scoring)
round(np.mean(score)*100, 2)
```

```python
from sklearn.model_selection import cross_val_predict

from sklearn.metrics import f1_score
train_pred = cross_val_predict(clf, train_data, train_label, cv=10, n_jobs=1)
f1_score(train_label,train_pred)
```

```python
from sklearn. metrics import roc_auc_score
roc_auc_score(train_label,train_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators=13)
score = cross_val_score(clf, train_data, train_label, cv=10, n_jobs=1, scoring=scoring)
round(np.mean(score)*100, 2)
```

```python
train_pred = cross_val_predict(clf, train_data, train_label, cv=10, n_jobs=1)
f1_score(train_label,train_pred)
```

```python
roc_auc_score(train_label,train_pred)
```

```python
from sklearn.svm import SVC

clf = SVC()
score = cross_val_score(clf, train_data, train_label, cv=10, n_jobs=1, scoring=scoring)
round(np.mean(score)*100, 2)
```

```python
train_pred = cross_val_predict(clf, train_data, train_label, cv=10, n_jobs=1)
f1_score(train_label,train_pred)
```
